```python
#W4A6

import math

a, b = map(int,input().split())

def ssnt(n):
    primes = [True] * (b+1)

    primes[0], primes[1] = False, False

    for i in range(2, int(math.sqrt(b+1)) + 1):
        if primes[i]:
            for j in range(i*i, b+1, i):
                primes[j] = False
    return [i for i in range(2, b+1) if primes[i]]

pr_b = ssnt(b)

tong = 0

for i in pr_b:
    if i < a:
        continue
    tong += i

print(tong)
```
