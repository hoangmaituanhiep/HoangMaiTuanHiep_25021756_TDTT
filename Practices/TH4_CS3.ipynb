```python
#W4A1

n = int(input())

tong = 0

if n%2 == 0:
    for i in range(1, (n//2)+1):
        tong += (n+1)
    print(tong)
else:
    i = 1
    while i <= n//2:
        tong += n
        i+=1
    print(tong+n)
```
