```python
#W4A8

n = int(input())

def reverse(n):
    tempt = 0

    while n >= 1:
        tempt = tempt*10 + n%10
        n = n//10

    return tempt

step = 0

while n != reverse(n):
    step += 1
    n = n + reverse(n)

print(step, n)
```
