```python
#W4A6

import math

a, b = map(int,input().split())

def ssnt(n):
    primes = [True] * (b+1)

    primes[0], primes[1] = False, False

    for i in range(2, int(math.sqrt(b+1)) + 1):
        if primes[i]:
            for j in range(i*i, b+1, i):
                primes[j] = False
    return [i for i in range(2, b+1) if primes[i]]

pr_b = ssnt(b)

tong = 0

for i in pr_b:
    if i < a:
        continue
    tong += i

print(tong)
```

```python
#W4A8

n = int(input())

def reverse(n):
    tempt = 0

    while n >= 1:
        tempt = tempt*10 + n%10
        n = n//10

    return tempt

step = 0

while n != reverse(n):
    step += 1
    n = n + reverse(n)

print(step, n)
```
