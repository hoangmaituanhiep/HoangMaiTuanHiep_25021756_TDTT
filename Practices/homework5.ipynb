```python
#W6A16
n = int(input())
m = int(input())
l = []

for i in range(n):
    line = list(map(int, input().split()))
    l.append(line)

tempt = []
for i in range(n):
    for j in range(m):
        tempt.append(l[i][j])
print(tempt)
size = (m*n)//2
print(f'[{tempt[:size]},{tempt[size:]}]')
```

```python
#W6A17
n=int(input())
l = [list(map(int, input().split())) for i in range(n)]

for i in range(n):
    print(l[i][i], end=', ')
print()
for i in range(n):
    print(l[i][n-i-1], end=', ')
```

```python
#W6A18
n,m,k = map(int,input().split())
l = [list(map(int, input().split())) for i in range(n)]
sum = 0

for i in range(n):
    sum += l[i][k-1]

print(sum)
```

```python
#W6A19
l = list(map(int,input().split()))

def maxl(arr):
    max_val = 0

    for i in arr:
        if max_val < i:
            max_val = i

    return max_val

print(maxl(l))
```

```python
#W6A20
l = list(map(int, input().split()))
k = int(input())

def find_k(arr, k):
    for i, v in enumerate(arr):
        if v == k:
            return i
    return -1

print(find_k(l,k))
```
