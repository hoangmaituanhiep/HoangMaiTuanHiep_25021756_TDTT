```python
#W6A16
n = int(input())
m = int(input())
l = []

for i in range(n):
    line = list(map(int, input().split()))
    l.append(line)

tempt = []
for i in range(n):
    for j in range(m):
        tempt.append(l[i][j])
print(tempt)
size = (m*n)//2
print(f'[{tempt[:size]},{tempt[size:]}]')
```

```python
#W6A19
l = list(map(int,input().split()))

def maxl(arr):
    max_val = 0

    for i in arr:
        if max_val < i:
            max_val = i

    return max_val

print(maxl(l))
```

```python
#W6A20
l = list(map(int, input().split()))
k = int(input())

def find_k(arr, k):
    for i, v in enumerate(arr):
        if v == k:
            return i
    return -1

print(find_k(l,k))
```

```python
#W6A23
l = list(map(int,input().split()))
x = int(input())

i = 0
n = len(l)

while i<n:
    if l[i] < x:
        l.pop(i)
        n-=1
    else:
        i+=1

print(l)
```

```python
#W6A26
l = list(map(int,input().split()))
n = len(l)
sum = 0

if n == 3:
    print(l[1])
elif n < 3: print("?")
else:
    for i in range(1, n-3):
        if l[i] < l[i+1] and l[i+1] > l[i+2]:
            sum += l[i+1]

    print(sum)
```

```python
#W6A27
l = list(map(int,input().split()))
x = int(input())
tempt = []

for i, v in enumerate(l):
    if v < x:
        tempt.append(i)

if len(tempt) > 0: print(tempt)
else: print(-1)
```
