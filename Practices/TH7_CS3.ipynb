```python
#1
def binary_search(arr, left, right, target) -> int:
    while left <= right:
        mid = (left+right)//2

        if arr[mid] == target:
            return mid

        if arr[mid] > target:
            right = mid - 1
        else:
            left = mid + 1

    return -1

def main():
    t = int(input())

    while t:
        l = list(map(int,input().split()))
        tar = int(input())

        print(f'{binary_search(l, 0, len(l)-1, tar)} \n')

        t-=1

if __name__ == "__main__":
    main()
```

```python
#2
def count_occurences(arr, x) -> int:
    cnt = 0
    
    for i in arr:
        if i==x:
            cnt+=1

    return cnt

def main():
    t = int(input())

    while t:
        arr = list(map(int,input().split()))
        tar = int(input())

        print(count_occurences(arr, tar))

        t-=1

if __name__ == "__main__":
    main()
```

```python
#3
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        check = False

        for j in range(n-i-1):
            if arr[j] >= arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                check = True

        if check == False:
            break;

    return arr

def main():
    l = list(map(int,input().split()))

    print(bubble_sort(l))

if __name__ == "__main__":
    main()
```

```python
#4
def find_mod(arr):
    times = {}
    for i in arr:
        times[i] = 0
    for i in arr:
        times[i] += 1

    max_times = 0
    ele = 0

    for key in times:
        if times[key] > max_times:
            ele = key
            max_times = times[key]

    print(f'{ele} : {max_times}')

def main():
    l = list(map(int,input().split()))
    find_mod(l)

if __name__ == "__main__":
    main()
```

```python
#5
arr = list(map(int,input().split()))
x = int(input())

n = len(arr)
cnt = 0

index = 0
while index<n:
    for i in range(index+1,n):
        if arr[index] + arr[i] == x:
            cnt += 1

    index+=1

print(cnt)
```

```python
#7
nums = list(map(int,input().split()))

n = len(nums)
size = 1

for i in range(n):
    cnt = 1
    
    tempt = i
    j = i+1

    while j<n:
        if nums[j] > nums[tempt]:
            cnt+=1
            tempt = j
        j+=1

    if cnt > size:
        size = cnt

print(size)
```
