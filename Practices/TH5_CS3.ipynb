```python
#W5A3 
import math

n = int(input())

def ssnt(n):
    primes = [True] * (n+1)

    primes[0] = primes[1] = False

    for i in range(2, math.isqrt(n) + 1):
        if primes[i]:
            for i in range(i*i, n+1, i):
                primes[i] = False
    return primes[n]

print(ssnt(n))
```

```python
#W5A4
n = int(input())

def check(n):
    sum = 0

    for i in range(1, n):
        if n%i == 0:
            sum += i
    return sum

if check(n) == n:
    print(True)
else: print(False)
```

```python
#W5A7
n1, n2 = map(int, input().split())
phep = input()

def maytinh(n1, n2, phep):
    if phep == '+':
        return n1 + n2
    if phep == '-':
        return n1 - n2
    if phep == '*':
        return n1 * n2
    if phep == '/':
        return f'{n1 / n2:.2f}'
print(maytinh(n1, n2, phep))
```

```python
#W5A8
x, y = map(int, input().split())

def binary(n):
    b = ''
    while n >= 1:
        if n%2 == 0:
            b = '0' + b
        else: b = '1' + b

        n//=2
    return b

bi_x = binary(x)
bi_y = binary(y)

nx, ny = len(bi_x), len(bi_y)

if nx > ny:
    bi_y = (nx-ny)*'0' + bi_y
else:
    bi_x = (ny-nx)*'0' + bi_x

dem = 0

for i in range(max(nx,ny)):
    if bi_x[i] != bi_y[i]:
        dem+=1

print(dem)

#7 = 111
#10 = 1010
```

```python
#W5A9
n = int(input())
def tong(x):
    tong = 0
    for i in range(1, 1000):
        tong += 1
        x //= 10
        if x == 0:
            break

    return tong

print(tong(n))
```

```python
#W5A10
a, b = input().split()

def check(a,b):
    move = {}
    
    for i in range(len(a)):
        move[a[i]] = b[i]
    
    for i in range(len(a)):
        if move[a[i]] != b[i]:
            return False
    return True

print(check(a,b))
```

```python
#1
n = int(input())

def check(n):
    size = len(str(n))
    tempt = 0

    while n:
        tempt += (n%10)**size
        n //= 10
    return tempt

print(check(n) == n)
```
