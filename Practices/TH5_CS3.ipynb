```python
#W5A3 
import math

n = int(input())

def ssnt(n):
    primes = [True] * (n+1)

    primes[0] = primes[1] = False

    for i in range(2, math.isqrt(n) + 1):
        if primes[i]:
            for i in range(i*i, n+1, i):
                primes[i] = False
    return primes[n]

print(ssnt(n))
```

```python
#W5A4
n = int(input())

def check(n):
    sum = 0

    for i in range(1, n):
        if n%i == 0:
            sum += i
    return sum

if check(n) == n:
    print(True)
else: print(False)
```

```python
#W5A5
sn = list(map(int, input().split()))
k = int(input())

def vitri(danhsach, n):
    for i in range(len(danhsach)):
        if danhsach[i] == k:
            return i
    return -1

print(vitri(sn, k))
```

```python
#W5A6
n = int(input())

def giaithua(n):
    if n <= 1:
        return 1
    return n*giaithua(n-1)

print(giaithua(n))
```
